```python
import pandas as pd
import matplotlib.pyplot as plt

camden_applications = 'Camden_Licensing_Applications_Beta.csv'
camden_business = 'Companies_Registered_In_Camden_And_Surrounding_Boroughs.csv'

#camden_applications = 'C:/Users/timothy/Downloads/Camden_Licensing_Applications_Beta.csv'
#camden_business = 'C:/Users/timothy/Downloads/Companies_Registered_In_Camden_And_Surrounding_Boroughs.csv'

cmdapp = pd.read_csv(camden_applications,parse_dates=['Received Date'])
cmdapp['Application Day'] = cmdapp['Received Date'].dt.weekday_name
```

```python
cmdapp['Application Day'].value_counts().plot(kind='bar', title='Total per Day')
plt.show()
```

```python
perward = cmdapp.groupby(['Ward Name'])


wardplot = []

for index, group in perward['Application Day']:
    plot = group.value_counts().sort_index().plot(kind='bar',title=index )
    wardplot.append(plot)
    plt.show()



    
```

```python
perward_avg = cmdapp.groupby(['Ward Name'])
perward_avg['Application Day'].count().plot(kind='bar', title='Total Applications per Ward')
plt.show()
```

```python
pertype = cmdapp.groupby(['Application Type'])
pertype['Application Reference'].count().plot(kind='bar')
plt.show()
```

```python
perwardtype = cmdapp.groupby(['Ward Name'])


for index, group in perwardtype['Application Type']:
    group.value_counts().plot(kind='bar',title=index)
    plt.show()
```
